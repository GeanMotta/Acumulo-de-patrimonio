```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import datetime as dt
```

```python
#Utilize as ferramentas abaixo para conseguir uma lista de datas para utilizar num dataframe gerado iterativamente
start = dt.datetime(2010, 1, 1)
end = dt.datetime(2019, 12, 30)

stnks = wb.DataReader("AAPL", "yahoo", start, end)

dates =[]
for x in range(len(stnks)):
    newdate = str(stnks.index[x])
    newdate = newdate[0:10]
    dates.append(newdate)
stnks['DATA'] = dates
stnks

#Este comando movera a coluna de data para a primeira coluna, facilitando a analise
colunas = stnks.columns.tolist()
colunas = colunas[-1:] + colunas[:-1]
colunas
stnks = stnks[colunas]



#Separa uma coluna para mes + ano
stnks['DATA'] = stnks['DATA'].astype('datetime64[ns]')
stnks["Month_Day"] = pd.DatetimeIndex(stnks['DATA']).month
stnks["Month_Day"] = stnks["Month_Day"].astype(str)
stnks["Year"] = pd.DatetimeIndex(stnks['DATA']).year
stnks["Year"] = stnks["Year"].astype(str)
stnks["MesEAno"] = stnks["Month_Day"] + "-" + stnks["Year"]
stnks = stnks.drop(columns=['Year', 'Month_Day'])

#Dataframe contendo somente as datas
as_datas = pd.DataFrame(stnks["DATA"])


#Dataframe contendo apenas datas de investimento e os primeiros dias
stnks = pd.DataFrame(stnks["MesEAno"])
stnks.drop_duplicates(["MesEAno"], keep="first", inplace= True)

stnks
```

```python
stnks.to_csv("Dates_Stocks.csv")
```
