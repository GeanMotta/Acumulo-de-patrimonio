```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas_datareader import data as wb
```

```python
stock = "AAPL"

df_older_prices = round(pd.DataFrame(wb.DataReader(stock, "yahoo", "2010-01-01", "2018-12-31")["Adj Close"]),4)
df_2019 = round(pd.DataFrame(wb.DataReader(stock, "yahoo", "2019-01-01", "2019-12-31")["Adj Close"]),4)
yearly_std = round(np.std(df_older_prices["Adj Close"]), 4)
```

```python
df_2019["Std.Dev. Up"] = df_2019["Adj Close"]["2019-01-02"] + yearly_std
df_2019["Std.Dev. Down"] = df_2019["Adj Close"]["2019-01-02"] - yearly_std
df_2019["Deviation check"] = (df_2019["Adj Close"] <= df_2019["Std.Dev. Up"]) & (df_2019["Adj Close"] >= df_2019["Std.Dev. Down"])
inside_confidence = round(((df_2019["Deviation check"] == True).sum() / df_2019["Deviation check"].count()),4)
```

```python
df_2019.plot(figsize=(16,9))
text_confidence = None
if inside_confidence >= 0.68:
    text_confidence = str("Results inside expected for 1 standart deviation, with {}% of the results inside confidence interval.".format(round(inside_confidence * 100, 2)))
else:
    text_confidence = str("Results outside expected for 1 standart deviation, with only {}% of the results inside confidence interval.".format(round(inside_confidence * 100, 2)))

print(text_confidence)
```
