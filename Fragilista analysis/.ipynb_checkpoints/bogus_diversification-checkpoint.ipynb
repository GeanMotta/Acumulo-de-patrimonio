```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as wb
```

```python
tickers=["AAPL", "AAPL.BA"]
cotacao = pd.DataFrame()
for t in tickers:
    cotacao[t]= wb.DataReader(t, data_source="yahoo", start="01-01-2018")[("Adj Close")]
```

```python
cotacao["Dólar CCL"] = ((cotacao["AAPL.BA"]*10)/(cotacao["AAPL"]))
cotacao.dropna()
```

```python
cotacao["AAPL DRC"] = cotacao["AAPL.BA"] / cotacao["Dólar CCL"]
```

```python
cotacao[["AAPL", "AAPL DRC"]].plot(figsize=(16,9))
plt.legend(fontsize=16)
plt.xlabel('Data', fontsize=18)
plt.ylabel('Valor em dólares', fontsize=16)
plt.savefig("BOGUS_DIVERSIFICATION.PNG")
```
