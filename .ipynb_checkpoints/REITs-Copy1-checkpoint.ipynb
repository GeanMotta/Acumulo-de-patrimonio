```python
import pandas as pd
from pandas_datareader import data as wb
import numpy as np
from matplotlib.pyplot import plot as plt
import datetime as dt
import math
```

```python
tickers = ["AMT", "CCI", "PLD", "EQIX", "SPG"]
start = dt.datetime(2010, 1, 1)
end = dt.datetime(2020, 9, 30)
```

```python
reits = pd.DataFrame()
for i in tickers:
    reits[i] = wb.DataReader(i, "yahoo", start, end)["Adj Close"]
```

```python
#Change investment variable accordingly
investment = 1000
reits = round(reits, 2)
```

```python
shares_bought = []
for i in tickers:
    shares_bought.append(math.floor(investment / reits[i][0]))
```

```python
shares_bought

accumulation = reits * shares_bought
```

```python
accumulation["Portfolio Value"] = accumulation.sum(axis=1)
```

```python
accumulation["Portfolio Value"].plot()
```
